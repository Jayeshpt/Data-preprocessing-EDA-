```python
# finding missing values
df.isnull().sum()
```

```python
# another way to find null values
[features for features in df.columns if df[features].isnull().sum()>0]
```

```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
# here we can see 'country code' in both dataset so we can combine these two columns
# 'df' is one dataset and 'df_country' is another one so merge two datasets
final_df=pd.merge(df,df_country,on='Country Code',how='left')
final_df
```

```python
# to check datatypes 
final_df.dtypes
```

```python
# how to find the number of datas in a columns is repeated
final_df.Country.value_counts()
```

```python
final_df.Cuisines.value_counts()
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_names
```

```python
country_values=final_df.Country.value_counts().values
country_values
```

```python
#here we can see country name and also getting values 
```

```python
plt.pie(country_values,labels=country_names)
```

```python
# here we can see the datas are messup and also some countries are in small deference 
# so its not possible to analyse properly so here ploting top 3 countries and its values
```

```python
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')
```

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text'])
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
# here .reset_index() helps to convert groupby data into dataframe
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()
```

```python
# here .rename(columns={0:'Rating count'}) used to change the column name 
rating=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
rating
```

### observation
1.when rating is 4.5 to 4.9 ---> excellent
2.when rating is 4.0 to 4.4 ---> very good
3.when rating is 3.5 to 3.9 ---> good
4.when rating is 2.5 to 3.4 ---> average
5.when rating is 1.8 to 2.4 ---> poor



```python
rating.head()
```

```python
import matplotlib as mlt
mlt.rcParams['figure.figsize']=(12,6) #this is used to change the parameter size
sns.barplot(x="Aggregate rating",y="Rating count",data=rating)
```

```python
# here we can plot ratings with given colors
import matplotlib as mlt
mlt.rcParams['figure.figsize']=(12,6) #this is used to change the parameter size
sns.barplot(x="Aggregate rating",y="Rating count",hue='Rating color',data=rating)
```

here we can see the given color is not matching so nee can give percular color as our wish

```python
# here we can plot ratings with perticular colors as given as dataset
import matplotlib as mlt
mlt.rcParams['figure.figsize']=(12,6) #this is used to change the parameter size
sns.barplot(x="Aggregate rating",y="Rating count",hue='Rating color',data=rating,palette=['blue','red','orange','yellow','green','green'])
```

```python
#frequency of rating points
sns.countplot(x="Rating color",data=rating,palette=['blue','red','orange','yellow','green','green'])
```

```python
# find out top most 10 cuisines

from collections import Counter
cuisines=final_df['Cuisines']
top_10 = Counter(cuisines).most_common(10)

print(top_10)
```
